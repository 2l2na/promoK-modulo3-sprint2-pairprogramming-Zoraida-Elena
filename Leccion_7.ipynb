1. ¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún
patrón evidente?

```python
# 'birth_rate'
```

```python
sns.histplot(x= 'birth_rate',            
             data = df, 
             bins = 20, 
             kde=True);
plt.xlabel('Tasa de natalidad')
plt.ylabel('')
```

```python
sns.boxplot(x= 'birth_rate', data = df);
```

2. ¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza
de vida? Proporciona una breve descripción de los patrones que observes.

```python
# 'life_expectancy' 'infant_mortality'
```

```python
sns.scatterplot(x = 'infant_mortality',
                 y='life_expectancy',
                 data = df,
                 size= 2,
                 color='red');
plt.xlabel('Mortalidad infantil')
plt.ylabel('Esperanza de vida')
```

```python
sns.regplot(x = 'infant_mortality',
                 y='life_expectancy',
                 data = df,               
                 color='red');
plt.xlabel('Mortalidad infantil')
plt.ylabel('Esperanza de vida')
```
